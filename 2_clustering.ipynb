# **CLUSTERING**

### Primero importamos librerías y el df a usar

```python
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('./data/retailmax.csv')
```

### Se elimina columna ID ya que no es funcional para análisis

```python
features = df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]
```

### Normalización y escalación de datos

```python
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
type(scaled_features)
```

### Ejecutamos algoritmo K-means para clustering de los datos, primero con 3 clusters

```python
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=42)
clusters = kmeans.fit_predict(scaled_features)
```

### Mostramos los clusters creados, categorías en las que se agruparon los datos

```python
clusters
```

```python
len(clusters)
```

### De nuevo pero con 4 clusters

```python
kmeans = KMeans(n_clusters=4, init='k-means++', max_iter=300, n_init=10, random_state=42)
clusters = kmeans.fit_predict(scaled_features)
```

```python
clusters
```

### Por último con 5 clusters

```python
kmeans = KMeans(n_clusters=5, init='k-means++', max_iter=300, n_init=10, random_state=42)
clusters = kmeans.fit_predict(scaled_features)
```

```python
clusters
```

```python
df
```

### Agregamos las categorías creadas por los clusters a cada dato en el df, en una nueva columna 'clasificación'

```python
df['clasificación'] = clusters
df
```

### Contamos cuantos valores hay en cada clasificación

```python
df['clasificación'].value_counts()
```

### Visualización de los resultados del clustering con un gráfico de dispersión

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Annual Income (k$)', y='Spending Score (1-100)', hue='clasificación', palette='Set1')
plt.title('Clusters de clientes')
plt.show()
```

### Calculamos la suma de cuadrados dentro de los clusters y graficamos con el método del codo

```python
    # Función para calcular la Suma de Cuadrados Dentro del Cluster (WCSS)
def calcular_wcss(datos):
    wcss = []
    for n in range(1, 11):
        kmeans = KMeans(n_clusters=n, init='k-means++', max_iter=300, n_init=10, random_state=42)
        kmeans.fit(datos)
        wcss.append(kmeans.inertia_)
    return wcss

# Calcular el WCSS para diferentes números de clusters
wcss = calcular_wcss(scaled_features)

# Graficar el método del codo
plt.plot(range(1, 11), wcss)
plt.title('Método del codo')
plt.xlabel('Número de clusters')
plt.ylabel('WCSS')
plt.show()
```

## ANÁLISIS DE RESULTADOS

### Análisis de los clusters

```python
cluster_analysis = df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)', 'clasificación']].groupby('clasificación').mean()
cluster_analysis
```

### Visualizamos información

```python
for cluster in cluster_analysis.index:
    print(f"Cluster {cluster}:")
    print(f" - Edad media: {cluster_analysis['Age'][cluster]:.2f}")
    print(f" - Ingreso anual medio: {cluster_analysis['Annual Income (k$)'][cluster]:.2f}k$")
    print(f" - Calificación RetailMax: {cluster_analysis['Spending Score (1-100)'][cluster]:.2f}")
    print("===========================================")
```

### Ahora, añadimos la variable Gender

```python
# Convertir a numérica
df['Gender'] = df['Gender'].map({'Male': 0, 'Female': 1})

# selección de features
extended_features = df[['Gender', 'Age', 'Annual Income (k$)', 'Spending Score (1-100)']]

# estandarización
scaled_extended_features = scaler.fit_transform(extended_features)

# Aplicar K-Means
kmeans = KMeans(n_clusters=5, init='k-means++', max_iter=300, n_init=10, random_state=42)
clusters_extended = kmeans.fit_predict(scaled_extended_features)

# Sustituir clusters anteriores
df['Cluster'] = clusters_extended

# análisis
extended_cluster_analysis = df[['Gender','Age', 'Annual Income (k$)', 'Spending Score (1-100)', 'Cluster']].groupby('Cluster').mean()
extended_cluster_analysis
```

### Y visualizamos resultados nuevos con la variable Gender 

```python
for cluster in extended_cluster_analysis.index:
    print(f"Cluster {cluster}:")
    print(f" - Edad media: {extended_cluster_analysis['Age'][cluster]:.2f}")
    print(f" - Ingreso anual medio: {extended_cluster_analysis['Annual Income (k$)'][cluster]:.2f}k$")
    print(f" - Calificación RetailMax: {extended_cluster_analysis['Spending Score (1-100)'][cluster]:.2f}")
    
    # porcentaje de mujeres
    porcentaje_mujeres = extended_cluster_analysis['Gender'][cluster] * 100
    print(f" - Porcentaje de mujeres: {porcentaje_mujeres:.2f}%")
    
    print("===========================================")
```
